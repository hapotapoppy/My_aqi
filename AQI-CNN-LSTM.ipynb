```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_csv("D:/My/Code/data/air-quality/柳州_空气质量_气象_处理后.csv")
feature_columns = ['PM10','PM2.5','O3','平均风速','风向_sin','NO2','10分钟平均能见度','CO','风向_cos','相对湿度','瞬时温度','SO2','地面气压']
target_column = 'AQI'
all_columns = feature_columns + [target_column]

scaler_feat = MinMaxScaler()
scaler_target = MinMaxScaler()
df[feature_columns] = scaler_feat.fit_transform(df[feature_columns])
df[target_column] = scaler_target.fit_transform(df[[target_column]])

def create_sequences(data, seq_len):
    X, y = [], []
    for i in range(len(data) - seq_len):
        X.append(data[feature_columns].iloc[i:i+seq_len].values)
        y.append(data[target_column].iloc[i+seq_len])
    return np.array(X), np.array(y)

seq_len = 24
X, y = create_sequences(df, seq_len)

train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
y_train = torch.FloatTensor(y_train).unsqueeze(1)
y_test = torch.FloatTensor(y_test).unsqueeze(1)

class AirQualityDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

batch_size = 32
train_dataset = AirQualityDataset(X_train, y_train)
test_dataset = AirQualityDataset(X_test, y_test)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

class OptimizedCNNLSTM(nn.Module):
    def __init__(self, input_features, cnn_out_channels, lstm_hidden_size, output_size):
        super(OptimizedCNNLSTM, self).__init__()
        
        self.cnn = nn.Sequential(
            nn.Conv1d(
                in_channels=input_features,  # 输入通道=特征数（8）
                out_channels=cnn_out_channels,  # 输出通道
                kernel_size=2,  # 核尺寸=2（适配24小时窗口，捕捉相邻2小时关联）
                stride=1,
                padding=1 
            ),
            nn.ReLU(),
            
        )
        self.lstm = nn.LSTM(
            input_size=cnn_out_channels,
            hidden_size=lstm_hidden_size,
            num_layers=2,
            batch_first=True,
            dropout=0.1
        )
        self.fc = nn.Linear(lstm_hidden_size, output_size)

    def forward(self, x):
        # x形状：(batch, seq_len, features) → 转置为(batch, features, seq_len)适配Conv1d
        x_trans = x.permute(0, 2, 1)
        # CNN输出：(batch, cnn_out_channels, seq_len) → 转置回(batch, seq_len, cnn_out_channels)适配LSTM
        cnn_out = self.cnn(x_trans).permute(0, 2, 1)
        # LSTM前向传播，取最后一个时间步输出
        lstm_out, _ = self.lstm(cnn_out)
        last_out = lstm_out[:, -1, :]
        output = self.fc(last_out)
        return output


input_features = len(feature_columns)
cnn_out_channels = 32  
lstm_hidden_size = 96
output_size = 1
epochs = 150 
lr = 0.001

model = OptimizedCNNLSTM(input_features, cnn_out_channels, lstm_hidden_size, output_size)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-5)

train_losses = []
test_losses = []

for epoch in range(epochs):
    model.train()
    train_loss = 0.0
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        train_loss += loss.item() * X_batch.size(0)
    train_loss /= len(train_loader.dataset)
    train_losses.append(train_loss)
    
    model.eval()
    test_loss = 0.0
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            test_loss += loss.item() * X_batch.size(0)
    test_loss /= len(test_loader.dataset)
    test_losses.append(test_loss)
    
    if (epoch + 1) % 5 == 0:
        print(f"Epoch {epoch+1}/{epochs}, Train Loss: {train_loss:.6f}, Test Loss: {test_loss:.6f}")


# 预测
model.eval()
with torch.no_grad():
    y_train_pred = model(X_train).numpy()
    y_train_true = y_train.numpy()
    y_test_pred = model(X_test).numpy()
    y_test_true = y_test.numpy()

# 反归一化
y_train_pred_raw = scaler_target.inverse_transform(y_train_pred.reshape(-1, 1))
y_train_true_raw = scaler_target.inverse_transform(y_train_true.reshape(-1, 1))
y_test_pred_raw = scaler_target.inverse_transform(y_test_pred.reshape(-1, 1))
y_test_true_raw = scaler_target.inverse_transform(y_test_true.reshape(-1, 1))

# 评估指标
train_mae = mean_absolute_error(y_train_true_raw, y_train_pred_raw)
train_mse = mean_squared_error(y_train_true_raw, y_train_pred_raw)
train_rmse = np.sqrt(train_mse)
train_r2 = r2_score(y_train_true_raw, y_train_pred_raw)

test_mae = mean_absolute_error(y_test_true_raw, y_test_pred_raw)
test_mse = mean_squared_error(y_test_true_raw, y_test_pred_raw)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test_true_raw, y_test_pred_raw)

print("\n===== 优化后模型评估指标 =====")
print(f"训练集: MAE={train_mae:.5f}, MSE={train_mse:.5f}, RMSE={train_rmse:.5f}, R²={train_r2:.5f}")
print(f"测试集: MAE={test_mae:.5f}, MSE={test_mse:.5f}, RMSE={test_rmse:.5f}, R²={test_r2:.5f}")


# 损失趋势
plt.figure(figsize=(10, 5))
plt.plot(range(1, epochs+1), train_losses, 'b-', label="训练损失", linewidth=2)
plt.plot(range(1, epochs+1), test_losses, 'r-', label="测试损失", linewidth=2)
plt.title("训练与测试损失变化趋势（优化后CNN-LSTM）", fontsize=14)
plt.xlabel("训练轮次", fontsize=12)
plt.ylabel("MSE损失", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# 测试集预测vs真实值
plt.figure(figsize=(12, 6))
plt.plot(y_test_true_raw[:500], label="真实AQI", color="blue", alpha=0.7)
plt.plot(y_test_pred_raw[:500], label="预测AQI", color="red", linestyle="--", alpha=0.7)
plt.title("AQI预测结果（优化后CNN-LSTM）", fontsize=14)
plt.xlabel("时间步", fontsize=12)
plt.ylabel("AQI值", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# 散点图
plt.figure(figsize=(8, 6))
plt.scatter(y_test_true_raw, y_test_pred_raw, alpha=0.5, color="green")
plt.plot([y_test_true_raw.min(), y_test_true_raw.max()], 
         [y_test_true_raw.min(), y_test_true_raw.max()], 
         'k--', lw=2, label="理想拟合线")
plt.title("真实AQI vs 预测AQI（优化后CNN-LSTM）", fontsize=14)
plt.xlabel("真实AQI", fontsize=12)
plt.ylabel("预测AQI", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
